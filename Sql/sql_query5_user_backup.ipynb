```python
import subprocess
```

```python
command = f"mysqldump --max_allowed_packet=64M -u root -p1234 sqldb > sqldb.sql"

# 백업 명령 실행
result = subprocess.run(command, shell=True, text=True)
```

```python
command = f"mysql -u root -p sqldb < sqldb.sql"

# 백업 명령 실행
result = subprocess.run(command, shell=True, text=True)
```
